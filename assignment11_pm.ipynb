Program 1 — Age Calculator (Improved Error Handling)

```python
age = int(input("Enter age: "))
print(f"In 10 years: {age + 10}")



while True:
    try:
        age = int(input("Enter age: "))

        if age < 0 or age > 150:
            raise ValueError(f"Age must be between 0 and 150, got {age}")

    except ValueError as e:
        print(f"Invalid input: {e}. Please enter a valid number.")

    else:
        print(f"In 10 years you will be {age + 10}")
        break

    finally:
        print("Age validation attempt complete.\n")
```

Program 2 — List Index Access

```python
numbers = [10, 20, 30, 40, 50]

index = int(input("Enter index: "))
print(numbers[index])
```

```python
numbers = [10, 20, 30, 40, 50]

while True:
    try:
        index = int(input("Enter index (0-4): "))

        if index < 0 or index >= len(numbers):
            raise IndexError(f"Index must be between 0 and {len(numbers)-1}")

    except ValueError:
        print("Invalid input. Please enter a numeric index.")

    except IndexError as e:
        print(f"Error: {e}")

    else:
        print(f"Value at index {index} is {numbers[index]}")
        break

    finally:
        print("Index check completed.\n")
```

Program 3 — Dictionary Lookup with Logging

```python
students = {"Amit": 85, "Neha": 90, "Rahul": 78}

name = input("Enter student name: ")
print(students[name])
```

```python
import logging

logging.basicConfig(
    filename="error_log.txt",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

students = {"Amit": 85, "Neha": 90, "Rahul": 78}

try:
    name = input("Enter student name: ")

    if name.strip() == "":
        raise ValueError("Name cannot be empty")

    if name not in students:
        raise KeyError(f"Student '{name}' not found")

except ValueError as e:
    print(f"Input Error: {e}")
    logging.error(e)

except KeyError as e:
    print("Student not found. Please check the name.")
    logging.error(e)

else:
    print(f"{name}'s marks: {students[name]}")

finally:
    print("Lookup attempt finished.")
```
